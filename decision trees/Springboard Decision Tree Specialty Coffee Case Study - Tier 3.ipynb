## Model 1: Entropy model - no max_depth

We'll give you a little more guidance here, as the Python is hard to deduce, and scikitlearn takes some getting used to.

Theoretically, let's remind ourselves of what's going on with a decision tree implementing an entropy model.

Ross Quinlan's **ID3 Algorithm** was one of the first, and one of the most basic, to use entropy as a metric.

**Entropy** is a measure of how uncertain we are about which category the data-points fall into at a given point in the tree. The **Information gain** of a specific feature with a threshold (such as 'spent_last_month <= 138.0') is the difference in entropy that exists before and after splitting on that feature; i.e., the information we gain about the categories of the data-points by splitting on that feature and that threshold. 

Naturally, we want to minimize entropy and maximize information gain. Quinlan's ID3 algorithm is designed to output a tree such that the features at each node, starting from the root, and going all the way down to the leaves, have maximial information gain. We want a tree whose leaves have elements that are *homogeneous*, that is, all of the same category. 

The first model will be the hardest. Persevere and you'll reap the rewards: you can use almost exactly the same code for the other models. 

```python
# Declare a variable called entr_model and use tree.DecisionTreeClassifier. 
entr_model = tree.DecisionTreeClassifier(criterion="entropy", random_state=1234)

# Call fit() on entr_model
entr_model.fit(X_train, y_train)

# Call predict() on entr_model with X_test passed to it, and assign the result to a variable y_pred 
y_pred = entr_model.predict(X_test)

# Call Series on our y_pred variable with the following: pd.Series(y_pred)
y_pred = pd.Series(y_pred)

# Check out entr_model
entr_model
```

```python
# Now we want to visualize the tree

# We can do so with export_graphviz
dot = tree.export_graphviz(
    entr_model,
    out_file=None,
    feature_names=X.columns,
    class_names= [str(c) for c in entr_model.classes_],
    filled=True,  
)

# Alternatively for class_names use entr_model.classes_
graph = pydotplus.graph_from_dot_data(dot)
Image(graph.create_png())
```

## Model 1: Entropy model - no max_depth: Interpretation and evaluation

```python
# Run this block for model evaluation metrics 
print("Model Entropy - no max depth")
print("Accuracy:", metrics.accuracy_score(y_test,y_pred))
print("Balanced accuracy:", metrics.balanced_accuracy_score(y_test,y_pred))
print('Precision score for "Yes"' , metrics.precision_score(y_test,y_pred, pos_label = "YES"))
print('Recall score for "Yes"' , metrics.recall_score(y_test,y_pred, pos_label = "YES"))
print('Precision score for "No"' , metrics.precision_score(y_test,y_pred, pos_label = "NO"))
print('Recall score for "No"' , metrics.recall_score(y_test,y_pred, pos_label = "NO"))
```

What can you infer from these results? Write your conclusions here:  

The entropy-based decision tree model, trained without a maximum depth constraint, achieved exceptionally high performance, with an accuracy of 99.16% and a balanced accuracy of 98.78%. This indicates strong overall predictive capability and well-balanced performance across both classes.

The model demonstrates excellent precision, achieving perfect precision for the "No" class (1.00) and very high precision for the "Yes" class (0.987). In terms of recall, the model correctly identifies all "Yes" instances (recall = 1.00), while recall for the "No" class is slightly lower (0.976), suggesting a small number of negative instances are misclassified.

Overall, the model is highly effective and shows a slight tendency toward prioritizing the correct identification of "Yes" cases, ensuring that no positive instances are missed.

However, the absence of a maximum depth constraint raises concerns about potential overfitting. The model may have learned patterns specific to the training data, which could reduce its ability to generalize to unseen data. Therefore, additional validation and the introduction of regularization techniques, such as limiting tree depth or pruning, are recommended to improve generalization and robustness.

## Model 2: Gini impurity model - no max_depth

Their equations are similar, but Gini impurity doesn't require logorathmic functions, which can be computationally expensive.

```python
# Run this block for model evaluation
print("Model Gini impurity model")
print("Accuracy:", metrics.accuracy_score(y_test,y_pred2))
print("Balanced accuracy:", metrics.balanced_accuracy_score(y_test,y_pred2))
print('Precision score for "YES"' , metrics.precision_score(y_test,y_pred2, pos_label = "YES"))
print('Precision score for "NO"' , metrics.precision_score(y_test,y_pred2, pos_label = "NO"))
print('Recall score for "NO"' , metrics.recall_score(y_test,y_pred2, pos_label = "NO"))
print('Recall score for "YES"' , metrics.recall_score(y_test,y_pred2, pos_label = "YES"))
```

How do the results here compare to the previous model? Write your judgements here: 

Both the Gini impurity and entropy-based decision tree models achieved very high performance, indicating that either criterion is effective for this classification task. However, there are some important differences in their results.

This indicates that the entropy model prioritized capturing all positive cases, even at the cost of a few false positives.

In contrast, the Gini model produced equal precision and recall values for both classes, indicating a more balanced distribution of errors. This suggests that the Gini model does not favor minimizing one type of error over another and maintains a more symmetrical classification behavior.

Overall, while the entropy model delivers slightly higher accuracy and ensures no missed "Yes" cases, it may be more prone to overfitting and bias toward the positive class. The Gini model, on the other hand, provides slightly lower but more balanced and stable performance.

In conclusion, the entropy model is preferable when maximizing detection of positive cases is critical, whereas the Gini model may be preferred for more balanced and consistent classification performance.

## Model 3: Entropy model - max depth 3
We're going to try to limit the depth of our decision tree, using entropy first.  

As you know, we need to strike a balance with tree depth. 

Insufficiently deep, and we're not giving the tree the opportunity to spot the right patterns in the training data.

Excessively deep, and we're probably going to make a tree that overfits to the training data, at the cost of very high error on the (hitherto unseen) test data. 

Sophisticated data scientists use methods like random search with cross-validation to systematically find a good depth for their tree. We'll start with picking 3, and see how that goes. 

```python
# Made a model as before, but call it entr_model2, and make the max_depth parameter equal to 3. 
# Execute the fitting, predicting, and Series operations as before
entr_model2 = tree.DecisionTreeClassifier(criterion="entropy", random_state=1234, max_depth =3)

# Call fit() on entr_model
entr_model2.fit(X_train, y_train)

# Call predict() on entr_model with X_test passed to it, and assign the result to a variable y_pred 
y_pred3 = entr_model2.predict(X_test)

# Call Series on our y_pred variable with the following: pd.Series(y_pred)
y_pred3 = pd.Series(y_pred3)

# Check out entr_model
entr_model2
```

```python
# As before, we need to visualize the tree to grasp its nature
dot3 = tree.export_graphviz(
    entr_model2,
    out_file=None,
    feature_names=X.columns,
    class_names= [str(c) for c in entr_model.classes_],
    filled=True,  
)

# Alternatively for class_names use entr_model2.classes_
graph = pydotplus.graph_from_dot_data(dot3)
Image(graph.create_png())
```

```python
# Run this block for model evaluation 
print("Model Entropy model max depth 3")
print("Accuracy:", metrics.accuracy_score(y_test,y_pred3))
print("Balanced accuracy:", metrics.balanced_accuracy_score(y_test,y_pred3))
print('Precision score for "Yes"' , metrics.precision_score(y_test,y_pred3, pos_label = "YES"))
print('Recall score for "Yes"' , metrics.recall_score(y_test,y_pred3, pos_label = "YES"))
print('Precision score for "No"' , metrics.precision_score(y_test,y_pred3, pos_label = "NO"))
print('Recall score for "No"' , metrics.recall_score(y_test,y_pred3, pos_label = "NO"))
```

So our accuracy decreased, but is this certainly an inferior tree to the max depth original tree we did with Model 1? Write your conclusions here: 

The entropy-based decision tree without a maximum depth achieved significantly higher performance, with an accuracy of 99.16% and a balanced accuracy of 98.78%, compared to the max depth 3 model, which achieved 90.76% accuracy and 86.59% balanced accuracy.

In terms of precision, the unrestricted model outperforms the max depth 3 model for the "Yes" class (0.987 vs 0.876), indicating that the simpler model produces more false positive predictions when identifying "Yes" instances. However, both models achieve perfect precision for the "No" class (1.00), meaning all predicted "No" cases are correct.

The max depth 3 model maintains perfect recall for the "Yes" class (1.00), ensuring no positive instances are missed, but at the cost of lower precision. Additionally, recall for the "No" class drops significantly (0.73), indicating that many negative instances are incorrectly classified as "Yes".

This suggests that the simpler model is biased toward predicting "Yes", prioritizing recall over precision for that class. While this reduces overall accuracy, it may still be useful in scenarios where missing positive cases is costly.

In conclusion, although the deeper entropy model achieves superior performance metrics, the max depth 3 model offers a simpler and more interpretable structure, albeit with reduced precision and balance. Therefore, it should not be considered inferior, but rather a trade-off between model complexity and generalization.

## 5. Random Forest
You might have noticed an important fact about decision trees. Each time we run a given decision tree algorithm to make a prediction (such as whether customers will buy the Hidden Farm coffee) we will actually get a slightly different result. This might seem weird, but it has a simple explanation: machine learning algorithms are by definition ***stochastic***, in that their output is at least partly determined by randomness. 

To account for this variability and ensure that we get the most accurate prediction, we might want to actually make lots of decision trees, and get a value that captures the centre or average of the outputs of those trees. Luckily, there's a method for this, known as the ***Random Forest***. 

Essentially, Random Forest involves making lots of trees with similar properties, and then performing summary statistics on the outputs of those trees to reach that central value. Random forests are hugely powerful classifers, and they can improve predictive accuracy and control over-fitting. 

Why not try to inform your decision with random forest? You'll need to make use of the RandomForestClassifier function within the sklearn.ensemble module, found [here](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html).
